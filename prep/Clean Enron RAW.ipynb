```python
path = r"F:\Documents\stopansko\masters\thesis\sig-detect\data\raw\enronraw\raw_inbox\maildir"
```

```python
emails = []
senders = []
```

```python
count = 0
for root, dirs, files in os.walk(path):
    if count > 5000:
        break
    if "\inbox" in root:
        for filename in files:
            count += 1
            try:
                with open(os.path.join(root, filename), encoding="utf-8") as f:
                    text = f.read()
                    msg = email.message_from_string(text)
                    emails.append(msg)
                    senders.append(msg["From"])
            except Exception as e:
                print(os.path.join(root, filename))
```

```python
from sklearn.model_selection import train_test_split
```

```python
def is_valid_sender(s):
    name, domain = tuple(s.split("@"))
    return all(x not in name for x in excluded_names) and all(x not in domain for x in excluded_domains)
```

```python
outside = [e for e in emails if is_valid_sender(e["From"]) and e["Content-Type"] and "text/plain" in e["Content-Type"]]
```

```python
words = [e for e in emails if "wishes" in e.get_payload().casefold() or "regards" in e.get_payload().casefold() ]
```

```python
clean_path = r"F:\Documents\stopansko\masters\thesis\sig-detect\data\clean\enron_raw"
for i, msg in enumerate(X_train):
    with open(os.path.join(clean_path, f"{i}.txt"), "w+", encoding="utf-8") as f:
        body = msg.get_payload()
        try:
            decoded = q.decodestring(body).decode("utf-8")
            f.write(decoded)
        except UnicodeDecodeError as ex:
            print("error")
```

```python
clean_path = r"F:\Documents\stopansko\masters\thesis\sig-detect\data\clean\enron_raw"
documents = []
for msg in X_train:
    body = msg.get_payload()
    try:
        decoded = q.decodestring(body).decode("utf-8")
        documents.append({"text": decoded, "labels": ["junk"]})
    except UnicodeDecodeError as ex:
        print("error")       
```

```python
import json
with open(os.path.join(clean_path, f"for_doccano.jsonl"), "w+", encoding="utf-8") as f:
    f.writelines("\n" + json.dumps(x, ensure_ascii=False) for x in documents)
```
